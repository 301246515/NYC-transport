```python
import dask.dataframe as dd
import pandas as pd
import numpy as np
import geopandas
from shapely.geometry import Point
from matplotlib import pyplot as plt
```

```python
df_in = dd.read_parquet('/data/green.parquet')
```

```python
df_in = df_in.reset_index()
```

```python
def spatial_join(df, lon_var, lat_var, locid_var):
    localdf = df[[lon_var, lat_var, locid_var]].copy()
    localdf[lon_var] = localdf[lon_var].fillna(value=0.)
    localdf[lat_var] = localdf[lat_var].fillna(value=0.)
    localdf['replace_locid'] = localdf[locid_var].isnull() & (localdf[lon_var] != 0.) & (localdf[lat_var] != 0.)
    
    if (np.any(localdf['replace_locid'])):
        shape_df = geopandas.read_file('../shapefiles/taxi_zones_latlon.shp')
        shape_df.drop(['OBJECTID', "Shape_Area", "Shape_Leng", "borough", "zone"],
                  axis=1, inplace=True)
        
        try:
            local_gdf = geopandas.GeoDataFrame(
                localdf, crs={'init': 'epsg:4326'}, 
                geometry=[Point(xy) for xy in 
                          zip(localdf[lon_var], localdf[lat_var])])
            local_gdf = geopandas.sjoin(local_gdf, shape_df, how='left', op='intersects')

            # restore values we do not want to replace
            local_gdf.LocationID.values[~local_gdf.replace_locid] = ((local_gdf[locid_var])[~local_gdf.replace_locid]).values
            return local_gdf.LocationID.rename(locid_var)
        except ValueError as ve:
            print(ve)
            return df[locid_var]
        
    else:
        return df[locid_var]


    

#     if np.any((lats[locs_to_replace] != 0.) | (lons[locs_to_replace] != 0.)):
#         points = [Point(xy) for xy in zip(lons, lats)]
#         points_df = geopandas.GeoDataFrame(
#             crs={'init': 'epsg:4326'}, geometry=points)
#         try:
#             joined = geopandas.sjoin(points_df, shape_df, op='intersects')
#             joined = joined.drop(['geometry', 'index_right'], axis=1)
            
#             # this seems like it could be done better
            
#             dflocal = (df[[locid_var, ]]).copy(True)
#             dflocal = dflocal.merge(joined, how='left', left_index=True,
#                                     right_index=True)
            
#             return dflocal.LocationID.rename(locid_var)
#         except ValueError as ve:
#             # this error occurs when there are no rows in joined due to
#             # no points having coordinates. Skip.
#             return df[locid_var]
#         except Exception as e:
#             print(e)
#             return df[locid_var]
#     else:
#         return df[locid_var]
```

```python
spatial_join(df_in.get_partition(0), "dropoff_longitude", "dropoff_latitude",
    "dropoff_location_id")
```

```python
df_in['dropoff_location_id'] = df_in.map_partitions(
    spatial_join, "dropoff_longitude", "dropoff_latitude",
    "dropoff_location_id", meta=('dropoff_location_id', np.float64))
df_in['pickup_location_id'] = df_in.map_partitions(
    spatial_join, "pickup_longitude", "pickup_latitude",
    "pickup_location_id", meta=('pickup_location_id', np.float64))
```

```python
df_in.to_parquet('/data/newgreen.parquet',
        compression="SNAPPY",
        has_nulls=True,
        object_encoding='json'                )
```
